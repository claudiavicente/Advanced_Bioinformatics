```python
# IMPORT THE NECESSARY LIBRARIES
import pandas as pd
import os
```

```python
# OPEN GDC MANIFEST
mf_name = '/Volumes/TOSHIBA EXT/4t curs/Advanced Bioinformatics/Project/Python/gdc_manifest.2023-09-19.txt'
data = pd.read_csv(mf_name, delimiter='\t')
```

```python
#CREATE THE FILE WHERE DF IS GOING TO BE SAVED
try:
    df_file_name = '/Volumes/TOSHIBA EXT/4t curs/Advanced Bioinformatics/Project/Python/expressiondata_id.csv'
    df_file = open(df_file_name, 'w')
    print('File has been created')
except Exception as e:
    print('File hasn\'t been opened:', e)
```

```python
# LOOPING THROUGH FILES IDs TO ACCESS THE FILES, DF IS CREATED WITH TPM EXPESSION DATA

compt = 0
df = pd.DataFrame()

for _, row in data.iterrows():
    file_name = str(row[1])
    
    file_path = '/Volumes/TOSHIBA EXT/4t curs/Advanced Bioinformatics/Project/Python/Files/' + file_name
    
    if os.path.exists(file_path):  
        data_sample = pd.read_csv(file_path, delimiter='\t', skiprows=[0, 2, 3, 4, 5])
        
        df[file_name] = data_sample.set_index('gene_id')['tpm_unstranded']
        compt += 1
        print(f'Sample nº {compt} added into df')
    else:
        print(f'File not found: {file_path}')
```

```python
df.to_csv(df_file_name)
df = df.loc[:, ~df.columns.duplicated()]
df.head(10)
```
