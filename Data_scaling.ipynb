```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
file_path = '/Volumes/TOSHIBA/4t/Advanced Bioinformatics/Project/Python/Rfiltered_data.csv' 
df = pd.read_csv(file_path, index_col = 0)
df
```

To scale the data, we will create a Standard Scaler instance. Then we will fit and transform the scaler on our data frame.

```python
scaler = StandardScaler().fit(df)
scaled_data = scaler.transform(df)
```

Scaled data has zero mean and unit variance:

```python
print(scaled_data.mean(axis = 0))
print(scaled_data.std(axis = 0))
```

```python
scaled_df = pd.DataFrame(scaled_data)
scaled_df = pd.DataFrame(data=scaled_data, index=df.index, columns=df.columns)
scaled_df.to_csv('/Volumes/TOSHIBA/4t/Advanced Bioinformatics/Project/Python/scaled_data.csv')
scaled_df
```
