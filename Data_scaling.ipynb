```python
import pandas as pd
from sklearn.preprocessing import RobustScaler
```

```python
file_path = '/Volumes/TOSHIBA EXT/4t curs/Advanced Bioinformatics/Project/Python/Rfiltered_data.csv'
df = pd.read_csv(file_path, index_col = 0)
df
```

To scale the data, we will create a Robust Scaler instance. Then we will fit and transform the scaler on our data frame.

```python
scaler = RobustScaler()
scaled_data = scaler.fit_transform(df)
```

```python
scaled_df = pd.DataFrame(scaled_data)
scaled_df = pd.DataFrame(data=scaled_data, index=df.index, columns=df.columns)
scaled_df.to_csv('/Volumes/TOSHIBA EXT/4t curs/Advanced Bioinformatics/Project/Python/scaled_data.csv')
scaled_df
```
